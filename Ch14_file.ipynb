## การอ่านเท็กซ์ไฟล์


````python
file = open('example.txt', 'r') # เปิดไฟล์ข้อความ
content = file.read() # อ่านเนื้อหาของไฟล์
file.close() # ปิดไฟล์

print(content) # พิมพ์เนื้อหาที่อ่านได้
````

## การเขียนเท็กซ์ไฟล์ที่ถูกเก็บด้วยรหัส Unicode


````python
# เขียนข้อมูลลงในไฟล์
file = open('hello.txt', 'w', encoding='utf-8')
file.write('Hello, World!')
file.write('ยินดีที่ได้รู้จัก')
file.write('บรรทัดที่ 3')
file.write('บรรทัดที่ 4')
file.close()
````

## การเขียนเท็กซ์ไฟล์พร้อมรหัสขึ้นบรรทัดใหม่


````python
# เขียนข้อมูลลงในไฟล์
file = open('hello2.txt', 'w', encoding='utf-8')
file.write('Hello, World!\n')
file.write('ยินดีที่ได้รู้จัก\n')
file.write('บรรทัดที่ 3\n')
file.write('บรรทัดที่ 4\n')
file.close()

# อ่านข้อมูลจากไฟล์ แล้วนำมาแสดงผลบนหน้าจอ
file = open('hello2.txt', 'r', encoding='utf-8')
print(file.read())
file.close()
````

## เทคนิคการเขียนเท็กซ์ไฟล์โดยใช้สตริงใน triple quote


````python
# เขียนข้อมูลลงในไฟล์
s = """Hello, World!
ยินดีที่ได้รู้จัก
บรรทัดที่ 3
บรรทัดที่ 4
"""
file = open("hello3.txt", "w", encoding="utf-8")
file.write(s)
file.close()

# อ่านข้อมูลจากไฟล์ แล้วนำมาแสดงผลบนหน้าจอ
file = open("hello3.txt", "r", encoding="utf-8")
print(file.read())
file.close()
````

## การอ่านไฟล์ `JSON` เข้ามาเป็น `dict`


````python
import json

with open('tourist_vocab.json', encoding='utf8') as f:
    vocabs=json.load(f)
print(f'type(vocabs) = {type(vocabs)}')
print(f'vocabs = {vocabs}')

print(f'vocabs["Hello"] = {vocabs["Hello"]}')
print(f'vocabs["Hello"]["meaning"] = {vocabs["Hello"]["meaning"]}')
````

## การอ่านไฟล์ `JSON` เข้ามาเป็น `list`


````python
import json

with open("drink.json", encoding="utf8") as f:
    drinks = json.load(f)
print(f"type(drinks) = {type(drinks)}")
print(f"drinks = {drinks}")
print()
print(f"drinks[0] = {drinks[0]}")
print(f'drinks[0]["menu"] = {drinks[0]["menu"]}')
print()
for drink in drinks:
    print(drink["menu"])
````

## การเขียนไฟล์ JSON 


````python
import json

d = [
    {'ไข่เจียวหมูสับ': [['ไข่', 3], ['หมู', 200]]},
    {'ต้มยำกุ้ง': [['กุ้ง', 7], ['ตะไคร้', 2], ['เห็ดฟาง', 5], ['พริกขี้หนู', 3]]}
]

# แปลง d ให้เป็นสตริงด้วย json.dumps()
j = json.dumps(d, ensure_ascii=False)
print(f"type(j) = {type(j)}")
print(j)

with open("ingredient.json", "w", encoding="utf-8") as f:
    f.write(j)
````

## การอ่านเท็กซ์ไฟล์จาก URL ด้วย `requests` 


````python
import requests

# URL ของไฟล์ที่ต้องการอ่าน
url = 'https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/gem.txt'

# ดาวน์โหลดไฟล์จาก URL
response = requests.get(url)

# ตรวจสอบสถานะการร้องขอ (Response status code)
if response.status_code == 200:
    # อ่านเนื้อหาของไฟล์
    content = response.text
    # พิมพ์เนื้อหาที่อ่านได้
    print(content)
else:
    print('Failed to retrieve the file.')
````

## การอ่านไฟล์รูปภาพจาก URL ด้วย `requests`

2. อ่านไฟล์ `spa.csv` และหาว่ามีลูกค้าที่นัดหมายในวันที่กำหนดหรือไม่

````python
filename = "spa.csv"
target_date = "2023-06-13"

with open(filename, 'r', encoding='utf-8') as file:
    customers = file.readlines()

for customer in customers:
    customer_data = customer.strip().split(",")
    appointment_date = customer_data[3]
    if appointment_date == target_date:
        print("ลูกค้าที่นัดหมายในวันที่", target_date, ":", customer_data[0])
        break
else:
    print("ไม่มีลูกค้าที่นัดหมายในวันที่", target_date)
````

4. เขียนโค้ดเพื่อนับว่ามีลูกค้าเลือกใช้บริการ `นวดผ่าเท้า` กี่ราย

````python
filename = "spa.csv"
target_service = "นวดผ่าเท้า"

with open(filename, 'r', encoding='utf-8') as file:
    customers = file.readlines()

count=0
for customer in customers:
    customer_data = customer.strip().split(",")
    service = customer_data[2]
    if service == target_service:
        count += 1
print(f"จำนวนลูกค้าที่เลือกบริการ{target_service}มีทั้งสิ้น {count} ครั้ง")
````
